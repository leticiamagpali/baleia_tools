### Calculating CG content

```python
from google.colab import files
from dataclasses import dataclass
from typing import Iterator

@dataclass
class SeqRecord:
  header: str
  sequence: str


  def __post_init__(self):
      if not self.header.startswith(">"):
          raise ValueError("Invalid FASTA header")

  @property
  def id(self):
    return self.header[1:].split()[0]

  @property
  def length(self):
    return len(self.sequence)

  @property
  def gc_content(self):
    seq = self.sequence.upper()
    return (seq.count("G") + seq.count("C")) / len(seq) if seq else 0

  @classmethod
  def read(cls, filepath: str) -> Iterator['SeqRecord']:
    """
    Generator: yields one FastaSeq object at a time from the file.
    """
    header = None
    seq_lines = []

    with open(filepath) as f:
        for line in f:
            line = line.strip()
            if not line:  # skip empty lines
                continue

            if line.startswith(">"):
                if header is not None:
                    yield cls(header, "".join(seq_lines))
                header = line
                seq_lines = []
            else:
                seq_lines.append(line)

        # yield last record
        if header is not None:
            yield cls(header, "".join(seq_lines))



fasta_file = files.upload()

for seq in SeqRecord.read("rosalind_gc.txt"):
  print(seq.id)
  print(seq.gc_content)


```
